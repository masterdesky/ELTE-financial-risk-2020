# Portfolio management and factors
## Assignment 3/1

Let $u \left( x \right)$ be a smooth utility function of a rational investor where $x$ denotes the value of financial wealth. We assume the investor maximizes the expected value of the utility function over the next period. For simplicity, we assume the initial wealth $x_{0} = 1$. Return of the investment portfolio, $r$, is a random variable, and we assume that there are a finite number of outcomes, each with probability $p_{i}$. The optimization problem is thus:

$$
\max \mathbb{E} \left[ u \left( 1 + r \right) \right]
=
\max \sum_{i} p_{i} u \left( 1 + r_{i} \right)
$$

Do a Taylor expansion around $1 + \mathbb{E} \left( r \right)$ up to the second order, and show that in this approximation, the investor has a mean-variance utility function.

## Solution

We define Taylor expansion as follows:

$$
T \left[ f \left( x \right) \right]
=
\sum_{n\,=\,0}^{\infty} \frac{f^{\left( n \right)} \left( a \right)}{n!} \left( x - a \right)^{n}
$$

Which up to the second order could be written in the following form:

$$
T \left[ f \left( x \right) \right]_{n\,=\,2}
=
f \left( a \right)
+
f' \left( a \right) \left( x - a \right)
+
\frac{f'' \left( a \right)}{2} \left( x - a \right)^{2}.
$$

Also by definition, variance of a random variable $X$ with mean $\mu$ is

$$
\sigma^{2}
=
\mathbb{E} \left[ \left( X - \mu \right)^{2} \right]
\equiv
\mathbb{E} \left[ \left( X - \mathbb{E} \left[ X \right] \right)^{2} \right]
$$

In the assignment we need to do the expansion around $1 + \mathbb{E} \left( r \right)$ of the $u \left( 1 + r \right)$ function. If we would Taylor expand it up to the fourth order we would get the mean-variance utility function, but expanding it to the second order is also good approximation, that's what we should do here.

The Taylor expansion is as follows:

$$
T \left[ u \left( 1 + r \right) \right]
\approx
u \left( 1 + \mathbb{E} \left[ r \right] \right)
+
u' \left( 1 + \mathbb{E} \left[ r \right] \right) \left( 1 + r - 1 - \mathbb{E} \left[ r \right] \right)
+
\frac{u'' \left( 1 + \mathbb{E} \left[ r \right] \right)}{2} \left( 1 + r - 1 - \mathbb{E} \left[ r \right] \right)^{2}
$$

$$
T \left[ u \left( 1 + r \right) \right]
\approx
u \left( 1 + \mathbb{E} \left[ r \right] \right)
+
u' \left( 1 + \mathbb{E} \left[ r \right] \right) \left( r - \mathbb{E} \left[ r \right] \right)
+
\frac{u'' \left( 1 + \mathbb{E} \left[ r \right] \right)}{2} \left( r - \mathbb{E} \left[ r \right] \right)^{2}
$$

Taking the expected value of this is the following:

$$
\mathbb{E} \left[ T \left[ u \left( 1 + r \right) \right] \right]
\approx
\mathbb{E} \left[ u \left( 1 + \mathbb{E} \left[ r \right] \right) \right]
+
\mathbb{E} \left[ u' \left( 1 + \mathbb{E} \left[ r \right] \right) \left( r - \mathbb{E} \left[ r \right] \right) \right]
+
\mathbb{E} \left[ \frac{u'' \left( 1 + \mathbb{E} \left[ r \right] \right)}{2} \left( r - \mathbb{E} \left[ r \right] \right)^{2} \right]
$$

The expected value of this smooth von Neumann-Morgenstern utility function $u \left( 1 + \mathbb{E} \left[ r \right] \right)$ and its derivates will be the same as the functions' values itself. Further simplifying the equation above gives us the following:

$$
\mathbb{E} \left[ u \left( 1 + r \right) \right]
\approx
u \left( 1 + \mathbb{E} \left[ r \right] \right)
+
u' \left( 1 + \mathbb{E} \left[ r \right] \right) \underbrace{\left( \mathbb{E} \left[ r \right] - \mathbb{E} \left[ r \right] \right)}_{0}
+
\frac{u'' \left( 1 + \mathbb{E} \left[ r \right] \right)}{2} \underbrace{\mathbb{E} \left[ \left( r - \mathbb{E} \left[ r \right] \right)^{2} \right]}_{\sigma^{2}}
$$

$$
\mathbb{E} \left[ u \left( 1 + r \right) \right]
\approx
u \left( 1 + \mathbb{E} \left[ r \right] \right)
+
\frac{u'' \left( 1 + \mathbb{E} \left[ r \right] \right)}{2} \sigma^{2}
$$

Which is a good approximation for the mean-variance utility function.
